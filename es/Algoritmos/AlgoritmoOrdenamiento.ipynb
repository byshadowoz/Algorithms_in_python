Los algoritmos de ordenamiento consisten en agarrar una lista de elementos desorganizada y ordenarla de menor a mayor

# Algoritmo de Ordenamiento Burbuja

Este algoritmo consiste en comparar dos elementos si el elemento de la izquierda es mayor al elemento de la derecha entonces intercambian posicion esto se va repetir hasta llegar al final de la lista donde nuevamente repetira el proceso hasta que todos los numeros de la izquierda sean menores a sus numeros de la derecha

## Problema 1
***
te dan una lista desorganizada y te piden que la organizes y ademas retornes el elemento mas grande de la lista (el ultimo)

```python
import random
lista = [random.randint(0,1000) for i in range(0,1000)]
```

```python
def burbuja(lista):
    # creamos una copia de la lista
    lista = list(lista)
    
    # vamos a repetir el proceso n-1 veces
    for _ in range(len(lista)-1):
        # vamos a iterar sobre la lista menos el ultimo elemento
        for i in range(len(lista)-1):
            # si el elemento actual es mayor que el siguiente, los intercambiamos
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]
    return lista
```

```python
burbuja(lista)
```

En el peor de los casos el algoritmo de burbuja tiene una complejidad de O(n^2)

# Algoritmo de ordenamiento por insercion

Es un algoritmo que funciona insertando iterativamente cada elemento en una lista no ordenada en su posicion correcta dentro de una parte ordenada de la lista 

```python
def insercion(nums):
    nums = list(nums)
    
    for i in range(0, len(nums)):
        act = nums[i] # guardamos el elemento en donde nos encontramos
        j = i-1 # definimos la variable j que el anterior de i 
        while j >= 0 and nums[j] > act: # mientras j sea mayor a cero y el elemento anterior sea mayor a nuestro elemento actual
            nums[j+1] = nums[j] # el elemento de la posicion j+1 es el element de la posicion j
            j -=1 #pasamos al numero anterior
        nums[j+1] = act # act pasa a ser el valor de la posicion j+1
    return nums
```

```python
insercion(lista)
```

al igual que el algoritmo de burbuja este tiene una complejidad de tiempo de O(n^2)

# Divide y conquista

***
Los siguientes algoritmos siguien una estrategia llamada dividir y conquistar, se refiere a descomponer un problema en dos o mas subproblemas relacionados hasta que sean lo suficientemente simpes para resolverse directamente para luego unir las soluciones de los subproblemas para obtener la solucion del problema original 

# Algoritmo de Ordenamiento Mezcla

el algoritmo de ordenamiendo Mezcla es necesario hacer un paso antes de crear el algoritmo en si, este algoritmo divide la lista en mitades y estas mitades en mas mitades si es necesario, cuando queda dos elementos estas necesitan otra funcion para que puedan ser comparadas 

## Merge

```python
def merge(nums1, nums2):
    # creamos una lista para almacenar los resultados
    merged = []
    
    #indices de iteracion
    i = 0
    j = 0
    
    # iteramos hasta que se acaben los elementos de alguna de las listas
    while i < len(nums1) and j < len(nums2):
        # el elemento mas pequeÃ±o es agregado a merged
        if nums1[i] <= nums2[j]:
            merged.append(nums1[i])
            i +=1
        else:
            merged.append(nums2[j])
            j +=1
    
    #tomamos las partes restantes
    nums1_cola = nums1[i:]
    nums2_cola = nums2[j:]
    
    # retornamos la lista final de resultados
    return merged + nums1_cola + nums2_cola
```

```python
def mezcla(lista):
    # terminamos si la lista es menor o igual a 1 elemento
    if len(lista) <= 1:
        return lista
    
    # obtenemos el punto medio
    mid = len(lista) // 2
    
    # dividimos la lista en dos partes
    izq = lista[:mid]
    der = lista[mid:]
    
    # resolvemos el problema recursivamente
    izq_ord, der_ord = mezcla(izq), mezcla(der)
    
    # combinamos los resultados de las dos partes
    num_ord = merge(izq_ord, der_ord)
    
    return num_ord
```

```python
mezcla(lista)
```

El algoritmo de Ordenacion Mezcla a pesar de que se vea muy grande tiene una complejidad menor a las anteriores siendo O(n log n) y una complejidad espacial de O(n)
